```python
import matplotlib.pyplot as plt
from random import randint
import math
import time

def serp_tri(pair):
    choice = randint(0,2)
    x = pair[0]
    y = pair[1]
    if choice == 0:
        x1,y1 = 0,0
    elif choice == 1:
        x1,y1 = .5, math.sqrt(1-.5**2)
    else:
        x1,y1 = 1,0
    return ((x+x1)/2,(y+y1)/2)


def experiment(pair):
    choice = randint(0,3)
    x = pair[0]
    y = pair[1]
    if choice == 0:
        x1,y1 = 0,0
    elif choice == 1:
        x1,y1 = 1,0
    elif choice == 2:
        x1,y1 = 0,1
    else:
        x1,y1 = 1,1
    return ((x+x1)/3,(y+y1)/3)
```

```python
func_list = [serp_tri]
start = time.time()

for func in func_list:
    my_list = [(0,0)]
    x_list = []
    y_list = []
    
    count = 0
    while count < 1000000-1:
        my_list.append(func(my_list[count]))
        count+=1

    for (a,b) in my_list:
        x_list.append(a)
        y_list.append(b)

    plt.figure(figsize=(20,20))
    plt.plot(x_list,y_list,'o',markersize=.1)
    plt.show()
    plt.savefig("Serpenski Triangle")
    plt.close()

    print(f"{len(my_list)} points")
    print(f"{time.time()-start:1.3f} seconds to complete")
```
